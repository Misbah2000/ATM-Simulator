```python
class ATM:

    def __init__(self,name,pin):
        self.name = name
        self.__pin = pin
        self.__balance = 0

    def check_balance(self):
        if self.check_pin():
            print(f"Your Current Balance is RS.{self.__balance}/-")

    def deposit(self):
        if self.check_pin():
            amount = int(input("Enter the Amount: "))
            if amount <=0:
                print("Please Enter a Valid Amount!")
            else:
                self.__balance += amount
                print(f"Rs.{amount}/- has been credited to your account.")
                print(f"Your Current Balance is Rs.{self.__balance}/-")

    def withdraw(self):
        if self.check_pin():
            amount = int(input("Enter the Amount: "))
            if amount <=0:
                print("Please Enter a Valid Amount!")
            elif amount > self.__balance:
                print("Insufficient Funds!")
            else:
                self.__balance -= amount
                print(f"Rs.{amount}/- has been credited to your account.")
                print(f"Your Current Balance is Rs.{self.__balance}/-")

    def change_pin(self):
        pass

    def check_pin(self):
        user_pin = int(input("Enter Your Pin: "))
        attempt = 1
        while True:
            if self.__pin == user_pin:
                return True
            else:
                print(f"Incorrect pin!, You are remaining {3-attempt} attempts left")
                if attempt >= 3:
                    print("Your Card is Blocked!")
                    break
                user_pin = int(input("Enter Your Pin: "))
                attempt += 1
    
```
